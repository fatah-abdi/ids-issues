```python
import numpy as np
import pandas as pd
mydf=pd.read_csv('//content/drive/MyDrive/dataset/CICIDS2017_sample_km.csv')
mydf[' Label'].value_counts().head(20)
```

```python
mydf_9 = mydf[(mydf[' Label']==2)]

mydf_a =mydf.drop(mydf_9.index)

mydf_b =mydf.drop(mydf_9.index)
```

```python
mydf_a=mydf_a.astype({" Label": int})
mydf_9=mydf_9.astype({" Label": int})
mydf_b=mydf_b.astype({" Label": int})
mydf_b=mydf_b.astype({" Label": int})
```

```python
mydf_b[' Label'].replace({0: "0", 1: "1", 2: "1", 3: "1", 4: "1", 5: "1", 6: "1", 7: "1", 8: "1", 9: "1", 10: "1", 11: "1", 12: "1", 13: "1"}, inplace=True)
mydf_a[' Label'].replace({0: "0", 1: "1", 2: "1", 3: "1", 4: "1", 5: "1", 6: "1", 7: "1", 8: "1", 9: "1", 10: "1", 11: "1", 12: "1", 13: "1"}, inplace=True)
mydf_9[' Label'].replace({0: "0", 1: "1", 2: "1", 3: "1", 4: "1", 5: "1", 6: "1", 7: "1", 8: "1", 9: "1", 10: "1", 11: "1", 12: "1", 13: "1"}, inplace=True)
mydf_b[' Label'].replace({0: "0", 1: "1", 2: "1", 3: "1", 4: "1", 5: "1", 6: "1", 7: "1", 8: "1", 9: "1", 10: "1", 11: "1", 12: "1", 13: "1"}, inplace=True)
```

```python
mydf_b=mydf_b.astype({" Label": int})
mydf_a=mydf_a.astype({" Label": int})
mydf_9=mydf_9.astype({" Label": int})
mydf_b=mydf_b.astype({" Label": int})
```

```python
X_train1 = mydf_a.drop([' Label'],axis=1) 
y_train1 = mydf_a.iloc[:, -1:].values.reshape(-1,1)
y_train1=np.ravel(y_train1)
X_test1 = mydf_9.drop([' Label'],axis=1) 
y_test1 = mydf_9.iloc[:, -1].values.reshape(-1,1)
y_test1=np.ravel(y_test1)
```

```python
X_train2 = mydf_b.drop([' Label'],axis=1) 
y_train2 = mydf_b.iloc[:, -1:].values.reshape(-1,1)
y_train2=np.ravel(y_train2)
X_test2 = mydf_9.drop([' Label'],axis=1) 
y_test2 = mydf_9.iloc[:, -1].values.reshape(-1,1)
y_test2=np.ravel(y_test2)
```

```python
X_train1=X_train1[[' Destination Port', ' Flow Duration', ' Total Length of Bwd Packets',
       'Bwd Packet Length Max', ' Bwd Packet Length Mean', ' Flow IAT Mean',
       ' Flow IAT Max', ' Fwd Header Length', 'Fwd Packets/s',
       ' Bwd Packets/s', ' Max Packet Length', ' Packet Length Mean',
       ' Packet Length Std', ' Packet Length Variance', ' Average Packet Size',
       ' Avg Bwd Segment Size', ' Fwd Header Length.1', ' Subflow Bwd Bytes',
       'Init_Win_bytes_forward', ' Init_Win_bytes_backward']]
X_test1=X_test1[[' Destination Port', ' Flow Duration', ' Total Length of Bwd Packets',
       'Bwd Packet Length Max', ' Bwd Packet Length Mean', ' Flow IAT Mean',
       ' Flow IAT Max', ' Fwd Header Length', 'Fwd Packets/s',
       ' Bwd Packets/s', ' Max Packet Length', ' Packet Length Mean',
       ' Packet Length Std', ' Packet Length Variance', ' Average Packet Size',
       ' Avg Bwd Segment Size', ' Fwd Header Length.1', ' Subflow Bwd Bytes',
       'Init_Win_bytes_forward', ' Init_Win_bytes_backward']]
```

```python
X_train2=X_train2[[' Destination Port', ' Flow Duration', ' Total Length of Bwd Packets',
       'Bwd Packet Length Max', ' Bwd Packet Length Mean', ' Flow IAT Mean',
       ' Flow IAT Max', ' Fwd Header Length', 'Fwd Packets/s',
       ' Bwd Packets/s', ' Max Packet Length', ' Packet Length Mean',
       ' Packet Length Std', ' Packet Length Variance', ' Average Packet Size',
       ' Avg Bwd Segment Size', ' Fwd Header Length.1', ' Subflow Bwd Bytes',
       'Init_Win_bytes_forward', ' Init_Win_bytes_backward']]
X_test2=X_test2[[' Destination Port', ' Flow Duration', ' Total Length of Bwd Packets',
       'Bwd Packet Length Max', ' Bwd Packet Length Mean', ' Flow IAT Mean',
       ' Flow IAT Max', ' Fwd Header Length', 'Fwd Packets/s',
       ' Bwd Packets/s', ' Max Packet Length', ' Packet Length Mean',
       ' Packet Length Std', ' Packet Length Variance', ' Average Packet Size',
       ' Avg Bwd Segment Size', ' Fwd Header Length.1', ' Subflow Bwd Bytes',
       'Init_Win_bytes_forward', ' Init_Win_bytes_backward']]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,precision_recall_fscore_support
from sklearn.metrics import f1_score,roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
dt = DecisionTreeClassifier(random_state = 0)
dt.fit(X_train2,y_train2 ) 
dt_score=dt.score(X_test2,y_test2)
y_predict=dt.predict(X_test2)
y_true=y_test2
print('Accuracy of DT: '+ str(dt_score))
precision,recall,fscore,none= precision_recall_fscore_support(y_true, y_predict, average='weighted') 


cm=confusion_matrix(y_true,y_predict)
f,ax=plt.subplots(figsize=(5,5))
sns.heatmap(cm,annot=True,linewidth=0.5,linecolor="red",fmt=".0f",ax=ax)
plt.xlabel("y_pred")
plt.ylabel("y_true")
plt.show()
```
